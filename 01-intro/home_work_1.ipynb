```python
# Q4. Missing values
# How many columns in the dataset have missing values? 
missing_cols = df.columns[df.isnull().any()]
print("Columns with missing values:", list(missing_cols))

num_cols_with_missing = df.isnull().any().sum()
print(f"Number of columns with missing values: {num_cols_with_missing}")

print("Number of Missing values per column")
print(df.isnull().sum().sort_values(ascending=False))
```

```python
# Q6. Median value of horsepower

# Find the median value of the horsepower column in the dataset.
original_median = df['horsepower'].median()
print("Original median horsepower:", original_median)

# Next, calculate the most frequent value of the same horsepower column.
most_freq = df['horsepower'].mode()[0]
print("Most frequent horsepower value:", most_freq)
    
# Use the fillna method to fill the missing values in the horsepower column with the most frequent value from the previous step.
df['horsepower_filled'] = df['horsepower'].fillna(most_freq)

# Now, calculate the median value of horsepower once again.
new_median = df['horsepower_filled'].median()
print("New median horsepower after filling missing values:", new_median)

# Has it changed? 
change = (new_median - original_median)
print("Yes, it changed by: +",change)
```

```python
# Q7. Sum of weights
# Select all the cars from Asia
asian_cars = df[df['origin'].str.lower() == 'asia']
print("snapshot of asian cars: ")
display(asian_cars.head(3))

# Select only columns vehicle_weight and model_year
vehicle_weight_n_mod_yr = df[['vehicle_weight','model_year']]
print("\nsnapshot of columns vehicle_weight and model_year: ")
display(vehicle_weight_n_mod_yr.head(3))
```

```python
# Select the first 7 values
    # Get the underlying NumPy array. Let's call it X.
    # Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
    # Invert XTX.
    # Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].
    # Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
    # What's the sum of all the elements of the result?

first_seven_records = vehicle_weight_n_mod_yr.head(7)
X = np.array(first_seven_records)
X_transpose = X.T
# multiplication of matrix : A @ B  --> matrix multiplication & A * B --> element-wise multiplication
XTX = X_transpose @ X
print("XTX is :",XTX)

# Invert XTX
XTX_inv = np.linalg.inv(XTX)
print("\nXTX_inv is :",XTX_inv)

# Array y 
y = [1100, 1300, 800, 900, 1000, 1100, 1200]
y = np.array(y)
print("\ny is: ",y)

# Compute w = (X^T X)^(-1) X^T y
w = XTX_inv @ X.T @ y

# Sum of all elements in w
w_sum = w.sum()

print("w:", w)
print("Sum of elements in w:", w_sum)
```
